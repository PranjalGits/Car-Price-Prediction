**Dataset mistakes:-**

- year has many non-numerical values and change the type of year from object to int.

- Price has one non numerical value and change the type of Price from object to int and also remove the commas.

- In kms_driven column we have to separate the values with integer and change the type of kms_driven from object to int and also clear the nan values and also remove the commas.

- fuel_type has nan values.

- company has many non meaning full rows.

- name column has many mixed type data so keep first 3 words of name.




**Cleaning the Data Set**

```python
backup = car.copy()
```

```python
car = car[car['year'].str.isnumeric()]     # remove non numeric data
```

```python
car['year'].isnull().sum()
```

```python
car['year'] = car['year'].astype(int)    # Change the datatype
```

```python
car.info()
```

```python
car = car[car['Price']!="Ask For Price"]     # remove non numeric data
```

```python
car['Price'].isnull().sum()
```

```python
car.info()
```

```python
car['Price']=car['Price'].str.replace(",","").astype(int)
```

```python
car.info()
```

```python
car['kms_driven']=car['kms_driven'].str.replace(",","").str.split(' ').str.get(0).astype(int)
```

```python
car=car[car['kms_driven'].str.isnumeric()]
```

```python
car.info()
```

```python
car=car[~car['fuel_type'].isna()]
```

```python
car.info()
```

```python
car['company'].unique()
```

```python
car['name']=car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
car
```

```python
car=car.reset_index(drop=True)
```

```python
car
```

```python
car.describe()
```

```python
car=car[car['Price']<6e6].reset_index(drop=True)
```

# **MODEL**

```python
##  Independent And dependent features
X=df.drop('Price',axis=1)
y=df['Price']
```

```python
X
```

```python
y.head()
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=661)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```python
ohe.fit(X[["name","company","fuel_type"]])
```

```python
ohe.categories_
```

```python
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae=mean_absolute_error(y_test,y_pred)
print("Mean absolute error", mae)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print("R2 Score", score)
```
